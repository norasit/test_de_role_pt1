### Question 3

Calculate total profit of "ข้าวโพดฝักอ่อน" produced in "ภาคเหนือ" from year 2555 to 2562
- profit rate in year 2555 - year 2559 is 50% (selling price is 50THB/kg.)
- profit rate in year 2560 - year 2562 is 30% (selling price is 120THB/kg.)
- assume all product was sold out every year.


```python
# Write you code and display question 3 result here.
df_corn_north = df[ (df['province'] == 'ภาคเหนือ')  & (df['product'] == 'ข้าวโพดฝักอ่อน') & (df['year'] >= 2555) & (df['year'] <= 2562) ]
df_corn_north
```

```python
# ฟังก์ชันสำหรับคำนวณ income
def calculate_income(row):
    # กำหนด selling price ตามช่วงปี
    if row['year'] <= 2559:
        selling_price = 50  # บาท/กิโลกรัม
    else:
        selling_price = 120  # บาท/กิโลกรัม
    
    # คำนวณรายได้ (income)
    income = row['production_ton'] * 1000 * selling_price
    return income

# ฟังก์ชันสำหรับคำนวณ profit 
def calculate_profit(row):
    # กำหนด profit rate และ selling price ตามช่วงปี
    if row['year'] <= 2559:
        profit_rate = 0.50
        selling_price = 50  # บาท/กิโลกรัม
    else:
        profit_rate = 0.30
        selling_price = 120  # บาท/กิโลกรัม
    
    # คำนวณกำไร (profit)
    profit = row['production_ton'] * 1000 * selling_price * profit_rate
    return profit

df_corn_north = df_corn_north.copy()
# ใช้ apply เพื่อคำนวณ profit และเพิ่มคอลัมน์ใหม่
# ใช้ apply เพื่อสร้าง column income และ profit_thb
df_corn_north['income'] = df_corn_north.apply(calculate_income, axis=1)
df_corn_north['profit_thb'] = df_corn_north.apply(calculate_profit, axis=1)
```

```python
df_corn_north
```

```python
# คำนวณกำไรรวมทุกปี
total_profit = df_corn_north['profit_thb'].sum()
print(f"Total profit from 2555 to 2562: {total_profit:,.2f} THB")
```

### Question 5

We plan to invest in growing soybeans (ถั่วเหลือง) which province should we invest.

We looking for
- high protential area (high havest_rate)
- we want market share only 20% but production_ton should be more than 2000 in the past (2550 - 2562)



```python
# เมื่อพิจารณา province นั้นมีรายชื่อภูมิภาคอยู่ด้วย
df['province'].unique()
```

```python
# สร้าง df_filter ที่เลือกเอา province ที่ไม่ใช่ชื่อภูมิภาคและไม่ใช่ "อื่นๆ
# รายชื่อ province ที่ต้องการกรองออก
exclude_provinces = ['ทั้งประเทศ', 'ภาคเหนือ', 'ภาคตะวันออกเฉียงเหนือ', 'ภาคกลาง', 'ภาคใต้', 'อื่นๆ']

# กรอง DataFrame ไม่เอา province ใน exclude_provinces
df_filtered = df[~df['province'].isin(exclude_provinces)]
```

```python
df_filtered['province'].unique()
```

```python

# Write you code and display question 5 result here.
df_soybeans = df_filtered[ df_filtered['product'] == 'ถั่วเหลือง']
```

```python
df_soybeans
```

```python
df_soybeans[ df_soybeans['province'] == 'ขอนแก่น' ]['production_ton'].sum()
```

```python
df_soybeans = df_soybeans.copy()

# 1. คำนวณ harvest_rate
df_soybeans['harvest_rate'] = df_soybeans['area_harvest_rai'] / df_soybeans['area_plant_rai']

# 2. คำนวณ production_ton รวมทั้งหมดของทุกจังหวัดในทุกปี (total_product_all)
total_product_all = df_soybeans['production_ton'].sum()
print(f"Total production (all provinces, all years): {total_product_all:,.2f} tons")

# 3. สร้าง DataFrame ใหม่ที่แสดงข้อมูลตามที่ต้องการ
result_df = (
    df_soybeans.groupby('province')
    .agg(
        avg_harvest_rate=('harvest_rate', 'mean'),      # ค่าเฉลี่ย harvest_rate
        total_production=('production_ton', 'sum')     # ผลรวม production_ton ของแต่ละจังหวัด
    )
    .reset_index()
)

# 4. คำนวณ market share percentage
result_df['market_share_percentage'] = (result_df['total_production'] / total_product_all) * 100

# แสดงผลลัพธ์
result_df

```

```python
result_df[ result_df['total_production'] > 2000 ].sort_values(by=['avg_harvest_rate', 'market_share_percentage'], ascending=[False, False])
```
